```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import time
```

```python
url = "https://www.formula1.com/en/results/2019/races"
page = urlopen(url)

soup = BeautifulSoup(page, "html.parser")
print(soup.prettify())
```

Finding each race-result by 'ui'

```python
for idx, each in enumerate(soup.find_all("ul")):
    print("["+str(idx)+"]"+"==================================")
    print(each.prettify())
```

soup.find_all("ul")[5] contains race-results.

```python
race_total_2019 = soup.find_all("ul")[5]

print(race_total_2019.prettify())
```

Race num starts from 1000 in href.

```python
a_list = race_total_2019.find_all(class_="block")

race_result = []
fastest_laps = []

url_2019 = []

for idx, a in enumerate(a_list):
    each_record = []
    race_result_url = a["href"]
    fastest_lap_url = str(a["href"]).replace("race-result", "fastest-laps")
    each_record.append("https://www.formula1.com/"+race_result_url)
    each_record.append("https://www.formula1.com/"+fastest_lap_url)
    url_2019.append(each_record)

url_2019 = url_2019[1:]
print(url_2019)
    
```

```python
for urls in url_2019:
    #print(urls)
    url_race_result = urls[0].replace("'","")
    url_fastest_lap = urls[1].replace("'","")
    

    response_rr = requests.get(url_race_result)
    response_rr.raise_for_status()

    response_fl = requests.get(url_fastest_lap)
    response_fl.raise_for_status()

   
    soup_2019_rr = BeautifulSoup(response_rr.text, "html.parser")
    soup_2019_fl = BeautifulSoup(response_fl.text, "html.parser")
    
    class_ = soup_2019_fl.find_all("tr", class_="bg-brand-white")[0]
    print(class_)
    # for each in tr:
    #     print(each.string)
    #print(class_)
```
